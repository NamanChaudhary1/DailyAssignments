# **Perform EDA on any of the dataset of your choice**

We Have chosen Data of Ipl matches 

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the IPL dataset
ipl_df = pd.read_csv('https://raw.githubusercontent.com/12345k/IPL-Dataset/master/IPL/data.csv')
```

## *Overview Of Dataset*

Display the first few rows of the dataset

```python
print(ipl_df.head())
```

Check the dimensions of the dataset

```python
print("Shape:", ipl_df.shape)
```

Get a summary of the dataset

```python
print(ipl_df.info())
```

## ***Exploratory data analysis and visualization:***

Explore the number of matches played in each season

```python
sns.countplot(x='season', data=ipl_df)
plt.title('Number of Matches Played in Each Season')
plt.xticks(rotation=90)
plt.show()
```

Explore the distribution of matches played in different cities

```python
sns.countplot(x='city', data=ipl_df)
plt.title('Matches Played in Different Cities')
plt.xticks(rotation=90)
plt.show()
```

Explore the number of matches won by each team

```python
sns.countplot(x='winner', data=ipl_df)
plt.title('Number of Matches Won by Each Team')
plt.xticks(rotation=90)
plt.show()
```

Explore the relationship between toss decision and match result

```python
sns.countplot(x='toss_decision', hue='result', data=ipl_df)
plt.title('Toss Decision vs Match Result')
plt.show()
```

Explore the distribution of player of the match awards

```python
player_of_match = ipl_df['player_of_match'].value_counts().head(10)
sns.barplot(x=player_of_match.index, y=player_of_match.values)
plt.title('Top 10 Players with Most Player of the Match Awards')
plt.xticks(rotation=90)
plt.show()
```

 Plot histogram for a numerical variable (e.g., 'Win_By_Runs')

```python
plt.figure(figsize=(10, 6))
plt.hist(ipl_df['win_by_runs'], bins=20, edgecolor='k')
plt.xlabel('Win By Runs')
plt.ylabel('Frequency')
plt.title('Histogram of Win By Runs')
plt.show()
```
